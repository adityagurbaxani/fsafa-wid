### Testing the hypotheses
#### Building Linear Regression Model

```python
import statsmodels.api as sm
```

```python
#Splitting dataframe into dependent and independent
x = df[['WID','POST','FE','WID*POST','WID*FE','POST*FE','WID*POST*FE','Size','Age','Leverage','ROA']]
y = df['EM']
```

```python
#Checking for colinearity
sns.pairplot(df[['EM','Size','Age','Leverage','ROA']])
```

```python
X = sm.add_constant(x)
wid = sm.OLS(y,X)
wid_res = wid.fit()
wid_res.summary()
```

#### Testing for Heteroskedasticity using Breusch-Pagan Test

```python
import statsmodels.stats.api as sms
from statsmodels.compat import lzip

bptest_result = sms.het_breuschpagan(wid_res.resid, wid_res.model.exog)
names = ['Lagrange multiplier statistic', 'p-value', 'f-value', 'f p-value']
lzip(names, bptest_result)
```

p-value < 0.05, indicates prescence of heteroskedasticity

#### Verifying the results using triple DiD

```python
did_df = df[['EM','WID','POST','FE','WID*POST','WID*FE','POST*FE','WID*POST*FE']]
did_df.reset_index(drop=True, inplace=True)

#Adding dummy records to adjust for WID=0 & FE=1 Scenario
em_FE_Pre = did_df[did_df['POST']==0].EM.mean()
em_FE_Post = did_df[did_df['POST']==1].EM.mean()
did_df.loc[len(did_df)] = [em_FE_Pre ,0,0,1,0,0,0,0] #Mean value of EM in pre-period as proxy
did_df.loc[len(did_df)] = [em_FE_Post,0,1,1,0,0,1,0] #Mean value of EM in post-period as proxy

did_df
```

```python
#Splitting into dependent and independent
did_x = did_df[['WID','POST','FE','WID*POST','WID*FE','POST*FE','WID*POST*FE']]
did_y = did_df['EM']
```

```python
#Running the Linear Regression
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(did_x, did_y)
```

```python
#print("Model Score:", model.score(did_x, did_y))

#Print the intercept and coefficents
print("b_%1d: %11.8f" % (0,model.intercept_))

for i in range(len(model.coef_)):
    print("b_%1d: %11.8f" % (i+1,model.coef_[i]))
```
